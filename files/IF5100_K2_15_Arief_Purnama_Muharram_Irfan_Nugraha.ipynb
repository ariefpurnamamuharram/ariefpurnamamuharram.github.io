# Quiz 2: Object Oriented Programming
___
Screenshot from 2022-10-06 22-07-54.png

```python
class BankAccount:
  def __init__(self, account_number, name, balance):
    super(BankAccount, self).__init__()
    self.account_number = account_number
    self.name = name
    self.balance = balance  
  def get_account_number(self):
    return self.account_number
  def get_name(self):
    return self.name
  def get_balance(self):
    return self.balance
  def set_name(self, name):
    self.name = name
  def set_balance(self, balance):
    self.balance = balance
  def deposit(self, money):
    self.balance += money
  def withdrawal(self, money):
    self.balance -= money
  def bank_fees(self):
    self.balance = self.balance - (self.balance * (5/100))
  def display(self):
    print('Account number:', self.account_number)
    print('Name:', self.name)
    print('Current balance:', self.balance)
```

Test the class.

```python
my_savings = BankAccount(23521013, 'Arief Purnama Muharram', 100000)
print('Current Saving Account')
print('-' * 13)
my_savings.display()

print('\n')

print('Deposit IDR50000')
print('-' * 13)
my_savings.deposit(50000)
my_savings.display()

print('\n')

print('Withdrawal IDR100000')
print('-' * 13)
my_savings.withdrawal(100000)
my_savings.display()

print('\n')

print('Apply Bank Fees')
print('-' * 13)
my_savings.bank_fees()
my_savings.display()
```
