b. Using the song “10 green bottles”, display the lines “There are [num] green bottles hanging
on the wall, [num] green bottles hanging on the wall, and if 1 green bottle should accidentally
fall”. Then ask the question “how many green bottles will be hanging on the wall?” If the user
answers correctly, display the message “There will be [num] green bottles hanging on the
wall”. If they answer incorrectly, display the message “No, try again” until they get it right.
When the number of green bottles gets down to 0, display the message “There are no more
green bottles hanging on the wall”.

```python
green_bottles = 10

print('There are ' + str(green_bottles) + ' green bottles hanging on the wall, ' + str(green_bottles) + ' green bottles hanging on the wall, and if 1 green bottle should accidentally fall.')

print('\n')

for i in range(green_bottles):
  while True:
    try_to_guess = int(input('How many green bottles will be hanging on the wall? '))
    if try_to_guess == green_bottles - (i + 1):
      if green_bottles - (i + 1) != 0:
        print('There will be ' + str(green_bottles - (i + 1)) + ' green bottles hanging on the wall')
      break
    else:
      print('No, try again')

print('There are no more green bottles hanging on the wall')
```

choose.

```python
dataset = {
    "John": {
        "N": 3056,
        "S": 8643,
        "E": 8441,
        "W": 2694
    },
    "Tom": {
        "N": 4832,
        "S": 6786,
        "E": 4737,
        "W": 3612
    },
    "Anne": {
        "N": 5239,
        "S": 4802,
        "E": 5820,
        "W": 1859
    },
    "Fiona": {
        "N": 3904,
        "S": 3645,
        "E": 8821,
        "W": 2451
    }
}

name = str(input('What is your name? '))
region = str(input('What is your region? '))

print('Your sales data is ', dataset[name][region])

check_update_request = int(input('Do you want to update your data? (1: Yes) '))
if check_update_request == 1:
  request_name = str(input('Please input your name: '))
  request_region = str(input('Please input your region: '))
  request_data = int(input('Please input your new sales number: '))
  dataset[request_name][request_region] = request_data
  print('Data have been updated successfully!')
  print('Your new sales data is', dataset[request_name][request_region])
```

**Function (Subprogram)**. Read this following instruction:

Define a subprogram that will ask the user to enter a number and save it as the variable
“num”. Define another subprogram that will use “num” and count from 1 to that number.
