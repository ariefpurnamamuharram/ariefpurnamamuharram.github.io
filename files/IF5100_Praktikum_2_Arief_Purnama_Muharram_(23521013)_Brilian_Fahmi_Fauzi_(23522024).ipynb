# Question 1
___
Write a function that can recursively sum the numbers in a list.<br>
Screenshot from 2022-09-08 22-03-25.png

```python
def sum_list(num_list):
  if len(num_list) == 1:
    return num_list[0]
  else:
    return num_list[0] + sum_list(num_list[1:])

print(sum_list([1,2,3]))

print(sum_list([1,2,3,-7]))
```

system that will be used as a practicum tool. Budi has a problem where he has to generate all
possible password sets of the system to use. Then, my friend suggested to create various
password sets with the words: “IF5100”. what are the passwords generated with that word

```python
availablePassword = [];

def generatePassword(word,i=0):
  if(i == len(word)):
    print(''.join(word))
  else:
    words = [c for c in word]
    for c in range(i,len(word)):
      words[c] , words[i] = words[i] , words[c]
      generatePassword(words,i+1)

print(generatePassword('IF5100'))
```

```python
def find_occurance(text_input, value_cari):
  if len(text_input) == len(value_cari):
    if text_input == value_cari:
      return 1
    else:
      return 0
  else:
    text_input = list(text_input)
    del text_input[0]
    text_input = ''.join(text_input)
    return find_occurance(text_input, value_cari)

find_occurance('mica', 'mic')
```
