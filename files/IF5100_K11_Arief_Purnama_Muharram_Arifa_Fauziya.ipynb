```python
#cek apakah terdapat upper case di dalam password
def check_upper(password):
  result = False
  for i in password:
    if i.isupper() == True:
      result = True
  return result

#cek apakah ada lower case di dalam password
def check_lower(password):
  result = False
  for i in password:
    if i.islower() == True:
      result = True
  return result

#cek apakah terdapat angka didalam password
def check_number(password):
  result = False
  for i in password:
    if i.isnumeric() == True:
      result = True
  return result

#cek apakah ada simbol di dalam password
def check_symbol(password):
  l_symbol = ['!', '$', '%', '&', '<', '*', '@', '£']
  result = False
  for i in password:
    if i in l_symbol:
      result = True
  return result

#cek kualitas password
def check_password_quality(password: str):
  password_score = 0
  if len(password) >= 8:
    password_score += 1
  if check_upper(password) == True:
    password_score += 1
  if check_lower(password) == True:
    password_score += 1
  if check_number(password) == True:
    password_score += 1
  if check_symbol(password) == True:
    password_score += 1
  return password_score

#membuat user baru
def create_new_userid(userid):
  if userid in l_users:
    print('Userid have been user. Please try the another one!')
    return False
  else:
    l_users.append(userid)

#membuat password baru
def create_password(userid):
  password = str(input('Insert your password: '))
  score = check_password_quality(password)
  if score <= 2:
    l_users.remove(userid)
    raise ValueError('Password too weak. Your account is rejected and have been deleted from the system!')
  elif score >= 3 and score <= 4:
    print('This password could be improved')
    check_try_again = str(input('Try again? (y: Yes) '))
    if check_try_again == 'y':
      create_password(userid)
    else:
      dict_users[userid] = {
          'password': password
      }
  else:
    dict_users[userid] = {
        'password': password
    }
    print('Account created!')

#untuk mengganti password
def change_password(userid):
  new_pass = str(input('Insert new password: '))
  if dict_users[userid]['password'] == new_pass:
    raise ValueError('Password can not same with the previous one!')
  else:
    score = check_password_quality(new_pass)
    if score <= 2:
      raise ValueError('Password change rejected!')
    elif score >= 3 and score <= 4:
      print('This password could be improved')
      check_try_again = str(input('Try again? (y: Yes) '))
      if check_try_again == 'y':
        change_password(userid)
      else:
        dict_users[userid]['password'] = new_pass
        print('Password have been changed!')
    else:
      dict_users[userid]['password'] = new_pass
      print('Password have been changed!')
```

```python
import datetime as dt
import random

class Account:
    def __init__(self,accountNumber,name,address,balance):
        self.accountNumber = accountNumber    # Variabel to Store accountNumber
        self.name = name                      # Variabel to Store name
        self.address = address                # Variabel to Store address
        self.balance = balance                # Variabel to Store balance


class ATMcard:
    def __init__(self,cardNumber,PIN,account):
        self.cardNumber = cardNumber    # Variabel to Store Card Number
        self.PIN = PIN                  # Variabel to Store PIN
        self.account = account          # Variable to store Account Object
        
```

```python
class Transaction:
    def __init__(self, transactionID, dateTime):
        self.transactionID = transactionID  # Variabel to Store transactionID
        self.dateTime = dateTime            # Variabel to Store Date

    def __str__(self):
        return f'Transaction[ transactionID = {self.transactionID}, dateTime = {self.dateTime}]'

# Modified by APM
class TransferTransaction(Transaction):
    def __init__(self, transactionID, dateTime,destinationNumber,amount):
      Transaction.__init__(self, transactionID, dateTime)
      self.amount = amount

    def __str__(self):
        return super(TransferTransaction, self).__str__() + f", TransferTransaction[ amount = {self.amount}]"


class WithdrawalTransaction(Transaction):
    def __init__(self, transactionID, dateTime,amount):
        super(WithdrawalTransaction, self).__init__(transactionID, dateTime)
        self.amount = amount


    def __str__(self):
        return super(WithdrawalTransaction, self).__str__() + f", WithdrawalTransaction[ amount = {self.amount}]"
```

```python
class ATMMachine:
    def __init__(self, kodeATM, cash):
        self.kodeATM = kodeATM            # ATM Code
        self.cash = cash                  # the amount of cash in the ATM
        self.isActive = False             # indication that ATM is being used or not (Active = True, !Active = False) 
        self.listOfTransactions = []      # List To Store Transaction Object
        self.currentCard = None           # Variabel to store ATM Card Object


    # Modified by APM
    def insertCard(self, card):
        '''
            set isActive to true and set currentCard to card
        '''
        self.isActive = True
        self.currentCard = card
 

    # Modified by APM
    def ejectCard(self):
        '''
            set isActive to dalse and set currentCard to nill
        '''
        self.isActive = False
        self.currentCard = None
 

    # Modified by APM
    def validatePIN(self,enteredPIN):
        '''
            This function aims to validate the entered pin

            check isActive(must be true)
            if active compare enteredPIN with currentCard PIN and return the result (boolean) else return false
        '''
        if self.isActive == False:
          raise ValueError('Insert card please!')
        
        if enteredPIN == self.kodeATM:
          return True
        else:
          return False


    # Modified by APM
    def seeBalance(self):
        '''
            this function aims to return the balance of the card

            Check isActive(must be true)
            if active return the account balance
        '''
        if self.isActive == False:
          raise ValueError('Insert card please!')

        return self.cash


    # Modified by APM
    def withdraw(self,withdrawAmount):
        '''
            this function aims to withdraw money from ATM

            check isActive(must be true)
            if active withdraw the desired amount of money from current account (balance)
            make sure the quantity of cash in ATM and current account balance sufficient
            if the condition is met, make sure the amount of cash in ATM and current account is reduced 
            then create new withdrawTransaction object and insert into listOfTransactions
        '''
        date = dt.datetime.now().strftime("%m/%d/%Y, %H:%M:%S")
        id = f'ID{random.randint(100,999)}'

        if self.isActive == False:
          raise ValueError('Insert card please!')
        
        if withdrawAmount > self.cash:
          raise ValueError('No sufficient balance!')
        
        self.cash -= withdrawAmount

        withdraw_transaction = WithdrawalTransaction(id, date, withdrawAmount)

        self.listOfTransactions.append(withdraw_transaction)
        

    # Modified by APM
    def transfer(self,destinationAccount, transferAmount):
        '''
            this function aims to send a certain amount of money to the destination account 

            check isActive(must be true)
            if active tranfer the desired amount of money from current account (balance) to destinationAccount
            make sure the quantity of cash in ATM and current account balance sufficient
            if the condition is met, make sure the amount of balance in current account is reduced and destination account balance increases 
            then create new transferTransaction object and insert into listOfTransactions
        '''
        date = dt.datetime.now().strftime("%m/%d/%Y, %H:%M:%S")
        id = f'ID{random.randint(100,999)}'

        if self.isActive == False:
          raise ValueError('Insert card please!')

        if transferAmount > self.cash:
          raise ValueError('No sufficient balance!')
        
        self.cash -= transferAmount
        destinationAccount.balance += transferAmount

        transfer_transaction = TransferTransaction(id, date, destinationAccount.accountNumber, transferAmount)

        self.listOfTransactions.append(transfer_transaction)

    
    def showAllTransactions(self):
        for lot in self.listOfTransactions:
            print(lot.__str__())
```
