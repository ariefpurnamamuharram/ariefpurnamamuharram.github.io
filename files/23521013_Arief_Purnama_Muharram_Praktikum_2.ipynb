## Regression model

### Predict sales for TV

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm


X_train, X_test, y_train, y_test = train_test_split(
    dataset['TV'], dataset['Sales'], test_size=0.3, random_state=0
)

x_train = sm.add_constant(X_train)
results = sm.OLS(y_train, x_train).fit()
print(results.summary())
```

Melalui *summary model* diperoleh nilai koefisien untuk TV adalah 0.0547 dan konstantanya adalah 7.1438. Dengan demikian, persamaan regresi yang terbantuk adalah **y = (0.0547 * TV) + 7.1438**.

### Predict sales for radio

### Predict sales for newspaper

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm


X_train, X_test, y_train, y_test = train_test_split(
    dataset['Newspaper'], dataset['Sales'], test_size=0.3, random_state=0
)

x_train = sm.add_constant(X_train)
results = sm.OLS(y_train, x_train).fit()
print(results.summary())
```

# Regresi Linear Berganda

## Import dataset

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


dataset = pd.read_csv('biaya_asuransi.csv')
```

## Pre-processing

```python
dataset.head(3)
```

```python
# Convert 'jenis_kelamin' variable
dict_jenis_kelamin = {'pria': 1, 'wanita': 2}
dataset['jenis_kelamin'] = dataset['jenis_kelamin'].apply(lambda x: dict_jenis_kelamin[x])

# Convert 'perokok' variable
dict_perokok = {'iya': 1, 'tidak': 0}
dataset['perokok'] = dataset['perokok'].apply(lambda x: dict_perokok[x])

# Convert 'area_tempat_tinggal' variable
dict_area_tempat_tinggal = {'barat_daya': 1, 'tenggara': 2, 'barat_laut': 3, 'timur_laut': 4}
dataset['area_tempat_tinggal'] = dataset['area_tempat_tinggal'].apply(lambda x: dict_area_tempat_tinggal[x])
```

## Describe dataset

```python
dataset.info()
```

```python
dataset.describe()
```

Hasil deskripsi data sebagai berikut:<br>

*   Terdapat 1338 baris data dalam dataset yang disediakan.
*   Umur paling muda pasien adalah 18 tahun, sedangkan paling tinggi adalah 64 tahun.
*   Indeks massa tubuh paling rendah dari pasien yang terdapat dalam dataset adalah 15.96, sedangkan tertinggi adalah 53.13.
*   Biaya kesehatan tertinggi yang dikeluarkan oleh pihak asuransi adalah 63770.428010, sedangkan terendah adalah 1121.873.
*   Biaya kesehatan rata-rata yang dikeluarkan oleh pihak asuransi adalah 13270.422265.



## Regression model

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm


X = dataset.drop(['biaya'], axis=1)
y = dataset['biaya']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)

x_train = sm.add_constant(X_train)
results = sm.OLS(y_train, x_train).fit()
print(results.summary())
```

Dengan demikian, 3 variabel yang memiliki nilai korelasi terbaik terhadap variabel 'biaya' adalah **umur**, **bmi**, dan **perokok**.

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm


X = dataset.drop(['jenis_kelamin', 'jumlah_anak', 'area_tempat_tinggal', 'biaya'], axis=1)
y = dataset['biaya']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)

x_train = sm.add_constant(X_train)
results = sm.OLS(y_train, x_train).fit()
print(results.summary())
```

Melalui *summary model* diperoleh nilai koefisien untuk umur adalah 260.3122, bmi 321.7844, perokok 2.344e+04, dan konstantanya adalah -1.169e+04. Dengan demikian, persamaan regresi yang terbantuk adalah **y = (260.3122 * umur) + (321.7844 * bmi) + (2.344e+04 * perokok) - 1.169e+04**.
