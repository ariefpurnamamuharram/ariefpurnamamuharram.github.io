# Search Word in Files
___
by Arief Purnama Muharram

```python
import os


def search_word(working_dir, word):
  founded_items = []

  def search(working_dir, word):
    print(f'\nOpen directory {working_dir}')
    items = os.listdir(working_dir)
    for item in items:
      if os.path.isdir(f'{working_dir}/{item}'):
        search(f'{working_dir}/{item}', word)
      else:
        print(f'> Reading file {item}')
        with open(f'{working_dir}/{item}', 'r') as fh:
          lines = fh.readlines()
          for line in lines:
            line = line.replace('\n', '').split(' ')
            if word in line:
              founded_items.append(item)
              break
  search(working_dir, word)

  print('\nWords found in:')
  for idx, item in enumerate(founded_items):
    print(f'{idx+1}. {item}')
```
