Data have not missing and duplicated values.

```python
# Preprocess 1: Object to int64
df_bus_jan['jumlah_penumpang'] = df_bus_jan['jumlah_penumpang'].apply(lambda x: int(x.replace(',', '')))
df_bus_feb['jumlah_penumpang'] = df_bus_feb['jumlah_penumpang'].apply(lambda x: int(x.replace(',', '')))
df_bus_mar['jumlah_penumpang'] = df_bus_mar['jumlah_penumpang'].apply(lambda x: int(x.replace(',', '')))
```

```python
# Visualize the number of passengers with the operation area "RUTE 14 (BLOK M - PD.LABU)" in each month.
passangers = []
passangers.append(['Januari', df_bus_jan[df_bus_jan['area_operasi'] == 'RUTE 14 (BLOK M - PD.LABU)']['jumlah_penumpang'].sum()])
passangers.append(['Februari', df_bus_feb[df_bus_feb['area_operasi'] == 'RUTE 14 (BLOK M - PD.LABU)']['jumlah_penumpang'].sum()])
passangers.append(['Maret', df_bus_mar[df_bus_mar['area_operasi'] == 'RUTE 14 (BLOK M - PD.LABU)']['jumlah_penumpang'].sum()])

df_passangers = pd.DataFrame(passangers, columns=['month', 'count'])

plt.figure(figsize=(10,5))
sns.barplot(data=df_passangers, x='month', y='count')
plt.show()
```

```python
# Visualize the number of passengers in each month.
passangers = []
passangers.append(['Januari', df_bus_jan['jumlah_penumpang'].sum()])
passangers.append(['Februari', df_bus_feb['jumlah_penumpang'].sum()])
passangers.append(['Maret', df_bus_mar['jumlah_penumpang'].sum()])

df_passangers = pd.DataFrame(passangers, columns=['month', 'count'])

plt.figure(figsize=(10,5))
sns.barplot(data=df_passangers, x='month', y='count')
plt.show()
```

There are 3850 missing values on the dataset. But, it's just okay.

```python
# Preprocessing 1: Market category
market_category = []
for item in df_car['Market Category'].dropna():
  categories = item.split(',')
  for category in categories:
    market_category.append(category)

df_market_category = pd.DataFrame(market_category, columns=['category'])
```

```python
# Preprocesisng 2: Car brands
car_brand = []
for brand in df_car['Make'].dropna().unique():
  car_brand.append([brand, df_car[df_car['Make'] == brand]['Make'].count()])

df_car_brand = pd.DataFrame(car_brand, columns=['brand', 'count'])
```

```python
# The market categories
plt.figure(figsize=(15,5))
sns.countplot(data=df_market_category, x='category')
plt.title('Market Categories')
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.pie(data=df_car_brand, x='count', labels='brand', autopct='%.0f%%')
plt.title('Car Brands')
plt.legend(title = "Brand:")
plt.show()
```

## Question 4 - Hotel
___

Screenshot from 2022-11-10 15-30-13.png

```python
df_hotel_bookings = pd.read_csv('hotel_bookings.csv')
df_hotel_bookings = df_hotel_bookings[df_hotel_bookings['arrival_date_year'] == 2016]
df_hotel_bookings.head(5)
```

```python
df_hotel_bookings.info()
```

```python
# Preprocessing 1: adults data
adults = []
for month in df_hotel_bookings['arrival_date_month'].dropna().unique():
  adults.append([month, df_hotel_bookings[df_hotel_bookings['arrival_date_month'] == month]['adults'].sum()])

df_adults = pd.DataFrame(adults, columns=['month', 'count'])
df_adults
```

```python
# Preprocessing 2: cancellations data
cancellations = []
for week in df_hotel_bookings['arrival_date_week_number'].dropna().unique():
  data = df_hotel_bookings[df_hotel_bookings['arrival_date_week_number'] == week]
  cancellations.append([week, data[data['is_canceled'] == 1]['is_canceled'].count()])
cancellations

df_cancellations = pd.DataFrame(cancellations, columns=['week', 'count'])
df_cancellations
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df_hotel_bookings, x='arrival_date_month')
plt.show()
```

```python
# Get the number of registered adults guests per month and sort by the biggest number
df_adults.sort_values(by='count', ascending=False)
```

```python
# Visualize the number of adult guests into diagram / chart that you think is appropriate
plt.figure(figsize=(15,5))
sns.barplot(data=df_adults, x='month', y='count')
plt.show()
```

```python
# Visualize the number of canceled bookings (is_cancelled, 0 = not cancelled, 1 = canceled) each week using the appropriate chart
plt.figure(figsize=(20,5))
sns.barplot(data=df_cancellations, x='week', y='count')
plt.show()
```
