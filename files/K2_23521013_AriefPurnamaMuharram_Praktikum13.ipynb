```python
from sklearn import datasets


iris = datasets.load_iris()

# Print the keys
print('Keys:', iris.keys())
```

```python
# Print the data description
print(iris.DESCR)
```

```python
# Print the data
print('Data:', iris.data)
```

```python
# Print the size of the data
print('Size:', len(iris.data))
```

```python
# Print the target
print('Target:', iris.target)
```

```python
# Print the target names
print('Target names:', iris.target_names)

# Print the feature names
print('Feature names:', iris.feature_names)
```

```python
# Print the iris dataset into a table-format style

import numpy as np
import pandas as pd


df = pd.DataFrame(np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
df
```

```python
# Split the dataset into the train and test dataset with test size 0.3.

from sklearn.model_selection import train_test_split


X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)

print('Train size:', len(X_train))
print('Test size:', len(X_test))
```

### KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier


knn = KNeighborsClassifier(n_neighbors=3)
knn = knn.fit(X_train, y_train)
preds = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report


print(classification_report(y_test, preds))
```

```python
import pandas as pd


df = pd.read_csv('car-sales-extended-missing-data.csv')
df.info()
```

```python
# Print the top-17 data
df.head(17)
```

```python
# Print the datatypes of each column of the dataset
df.dtypes
```

```python
# Count the missing values
df.isna().sum()
```

```python
# Remove rows with no labels (NaN's in the Price column)

df = df.dropna(subset=['Price'])

df
```

```python
# Prefer to clean the data that contain missing values
df = df.dropna(how='any')

df
```

```python
from sklearn.model_selection import train_test_split


X = df.drop(columns=['Price'])
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python
# Feature engineering

from sklearn.preprocessing import LabelEncoder


le_make = LabelEncoder()
le_make.fit(X['Make'])
X_train['Make'] = le_make.transform(X_train['Make'])
X_test['Make'] = le_make.transform(X_test['Make'])

le_colour = LabelEncoder()
le_colour.fit(X['Colour'])
X_train['Colour'] = le_colour.transform(X_train['Colour'])
X_test['Colour'] = le_colour.transform(X_test['Colour'])
```

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression


reg = LinearRegression()
reg = reg.fit(X_train, y_train)
preds = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error


print('MSE:', mean_squared_error(y_test, preds))
```

```python
# Print the first 15% data predictions
how_many = round(0.15 * len(X_test))
data = X_test.head(how_many)
print(reg.predict(data))
```

```python
import pandas as pd 


df = pd.read_csv('data_train.csv')
```

```python
df.describe()
```

```python
# Drop rows that contain missing values
df = df.dropna(how='any')
```

```python
# Feature engineering

from sklearn.preprocessing import LabelEncoder


le_gender = LabelEncoder()
le_gender.fit(df['Gender'])
df['Gender'] = le_gender.transform(df['Gender'])

le_employed = LabelEncoder()
le_employed.fit(df['Employed'])
df['Employed'] = le_employed.transform(df['Employed'])

le_transport = LabelEncoder()
le_transport.fit(df['Transport'])
df['Transport'] = le_transport.transform(df['Transport'])
```

```python
# Visualize the correlation matrix

import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Visualize the target label

import matplotlib.pyplot as plt
import seaborn as sns


plt.figure()
sns.countplot(data=df, x='Transport')
plt.title('Target Label Distribution')
plt.show()
```

### DecisionTree Model

```python
# Normalize the data using StandardScaler

from sklearn.preprocessing import StandardScaler

X = df.drop(columns=['Transport'])
y = df['Transport']

scaler = StandardScaler()
scaler = scaler.fit(X)
X = scaler.transform(X)
```

```python
# Split the data into the train and test dataset

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python
from sklearn.tree import DecisionTreeClassifier


clf = DecisionTreeClassifier(criterion='gini')
clf = clf.fit(X_train, y_train)
preds = clf.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import sklearn.tree as t


plt.figure(figsize=(25,25))
t.plot_tree(clf)
plt.title('Decision Tree')
plt.show()
```

```python
from sklearn.metrics import classification_report


print(classification_report(y_test, preds))
```
