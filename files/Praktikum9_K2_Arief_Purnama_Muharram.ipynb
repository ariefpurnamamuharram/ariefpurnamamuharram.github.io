NumPy copy. 'Copy' is a new array thus _owns_ the data and any changes made to the copy will not affect the original array, and vice versa.

```python
import numpy as np


arr = np.array([1, 2, 3, 4, 5])
x = arr.copy()

# Try to modify the original data
arr[0] = 42

print(f'arr: {arr}')
print(f'x: {x}')
```

NumPy view. 'View' is just a view of the original array thus any changes made to the view will affect the original array, and vice versa.

```python
import numpy as np


arr = np.array([1, 2, 3, 4, 5])
x = arr.view()

# Try to modify the original data
arr[0] = 42

print(f'arr: {arr}')
print(f'x: {x}')
```

Join two or more arrays in a single array with 'join'.

```python
import numpy as np


arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

arr = np.concatenate((arr1, arr2))

print(arr)
```

```python
# Join in 2-D array
import numpy as np


arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

arr = np.concatenate((arr1, arr2), axis=1)

print(arr)
```

```python
# Join arrays using stack().
# Stack have the same function as join, except the stacking is done along a new axis
import numpy as np


arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

arr = np.stack((arr1, arr2), axis=1)

print(arr)
```

```python
# Join arrays using hstack().
# hstack() helps to stack along rows.
import numpy as np


arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

arr = np.hstack((arr1, arr2))

print(arr)
```

```python
# Join arrays using vstack().
# vstack() helps to stack along columns.
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

arr = np.vstack((arr1, arr2))

print(arr)
```

```python
import numpy as np


def calculate(VelAsc, VelDsc, alfaAsc, alfaDsc, thetaAsc, thetaDsc):
  rad_alfaAsc = np.radians(alfaAsc)
  rad_alfaDsc = np.radians(alfaDsc)
  rad_thetaAsc = np.radians(thetaAsc)
  rad_thetaDsc = np.radians(thetaDsc)

  a = np.array([
      [VelAsc],
      [0],
      [VelDsc]
  ])
  b = np.array([
      [-1 * np.sin(rad_thetaAsc) * np.cos(rad_alfaAsc), np.sin(rad_thetaAsc) * np.cos(rad_alfaAsc), np.cos(rad_alfaAsc)],
      [-1, np.cos(rad_thetaAsc) * np.cos(np.radians(90 - thetaAsc)), 0],
      [-1 * np.sin(rad_thetaDsc) * np.cos(rad_alfaDsc), np.sin(rad_thetaDsc) * np.sin(rad_alfaDsc), np.cos(rad_thetaDsc)]
  ])

  result = np.matmul((np.linalg.inv(b)), (a))

  return result
```

```python
for idx, item in enumerate(dataset):
  result = calculate(item[2], item[3], item[4], item[5], item[6], item[7])
  print(f'Result #{idx+1}: \n{result}\n')
```
