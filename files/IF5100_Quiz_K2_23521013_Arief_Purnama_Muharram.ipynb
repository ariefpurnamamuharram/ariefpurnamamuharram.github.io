```python
import math


class Shape:
  def __init__(self, name, area, circumference):
    self._name = name
    self._area = area
    self._circumference = circumference
  def get_name(self):
    return self._name
  def set_name(self, name):
    self._name = name
  def get_area(self):
    return self._area
  def set_area(self, area):
    self._area = area
  def get_circumference(self):
    return self._circumference
  def set_circumference(self, circumference):
    self._circumference = circumference


class Circle(Shape):
  def __init__(self, radius, name='Circle'):
    Shape.__init__(self, name, 0, 0)
    self._radius = radius
    self.set_area(math.pi * radius * radius)
    self.set_circumference(2 * math.pi * radius)
  def cal_area(self, radius):
    return math.pi * radius * radius
  def cal_circumference(self, radius):
    return 2 * math.pi * radius
  def update(self, radius):
    self._radius = radius
    self.set_area(math.pi * self._radius * self._radius)
    self.set_circumference(2 * math.pi * self._radius)
  def __str__(self):
    return 'Shape name : ' + str(self._name) + ' with Radius : ' + str(self._radius) + ' has the Area of: ' + str(self._area) + ' units and Circumference : ' + str(self._circumference) + ' units'


class Square(Shape):
  def __init__(self, side, name='Square'):
    Shape.__init__(self, name, 0, 0)
    self._side = side
    self.set_area(side * side)
    self.set_circumference(4 * side)
  def cal_area(self, side):
    return side * side
  def cal_circumference(self, side):
    return 4 * side
  def update(self, side):
    self._side = side
    self.set_area(self._side * self._side)
    self.set_circumference(4 * self._side)
  def __str__(self):
    return 'Shape name : ' + str(self._name) + ' with Side : ' + str(self._side) + ' has the Area of: ' + str(self._area) + ' units and Circumference : ' + str(self._circumference) + ' units'
```
