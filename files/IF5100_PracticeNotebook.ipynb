```python
# [Shopping Days Problem]

import calendar as cal


def shopping_days(year):
  """
  year: a number <= 1941
  returns the number of days between US Thanksgiving and Christmas of the year
  """
  month = cal.monthcalendar(year,11)
  if month[0][cal.THURSDAY] != 0:
    thanksgiving = month[3][cal.THURSDAY]
  else:
    thanksgiving = month[4][cal.THURSDAY]
  num_of_shopping_days = (cal.monthrange(year,11)[1] - thanksgiving) + 25
  return num_of_shopping_days

print("In 1939", "the number of shopping days is", shopping_days(1939), ".")
```

```python
# [Read and Write Problem]

def create_collection(number_range=10):
  collections = []
  for i in range(number_range):
    collections.append(i)
  return collections

def write_to_file(collections, fn='intpos_file.txt'):
  '''
  collections: List of collection to be saved in the file
  fn: Filename, the default is intpos_file.txt
  Return none
  '''
  with open(fn, 'w') as fh:
    for i in collections:
      fh.write(str(i) + '\n')
  return None

def read_file(fn='intpos_file.txt'):
  '''
  fn: Filename, the default is intpos_file.txt
  Return none
  '''
  with open(fn, 'r') as fh:
    item = fh.read()
    print(item)
  return None

# Step 1: Create integer collections
collections = create_collection(10)

# Step 2: Write collections into a file
write_to_file(collections, fn='intpos_file.txt')

# Step 3: Read the saved file
read_file(fn='intpos_file.txt')
```

```python
# Iteration problem #1

def remove_dups(L1,L2):
  for e in L1:
    if e in L2:
      L1.remove(e)

L1 = [1, 2, 3, 4]
L2 = [1, 2, 5, 6]
remove_dups(L1,L2)

print('L1', L1)
print('L2', L2)
```

```python
# Iteration problem #1: Solution

def remove_dups(L1,L2):
  L1_copy = L1[:]
  for e in L1_copy:
    if e in L2:
      L1.remove(e)

L1 = [1, 2, 3, 4]
L2 = [1, 2, 5, 6]
remove_dups(L1,L2)

print('L1', L1)
print('L2', L2)
```

```python
# [Recursive Method]

def mult(a, b):
  if b == 1:
    return a
  else:
    return a + mult(a, b-1)

mult(2, 4)
```

```python
# [Recursive Method]
# Base case kasus n-factorial adalah 1.
# Karena 1! sudah tidak dapat dipecah kembali.

def factorial(n):
  if n == 1:
    return 1
  else:
    return n*factorial(n-1)

factorial(5)
```

```python
# IsOdd, a function to check whether an integer >= 0 is an odd number or not
# base = 0 -> false

def check_is_odd(n):
  if n == 1:
    return True
  elif n <= 0:
    return False
  else:
    return check_is_odd(n-2)

check_is_odd(11)
```

```python
# [1, 2, [3,4], [5,6]]

def list_sum(num_list):
  total = 0
  for element in num_list:
    if type(element) == list:
      total = total + list_sum(element)
    else:
      total = total + element
  return total

list_sum([1, 2, [3,4], [5,6]])
```

```python
class Fraction:
  def __init__(self, nominator, denominator):
    self.nominator = int(nominator)
    self.denominator = int(denominator)
  def get_nominator(self):
    return self.nominator  
  def get_denominator(self):
    return self.denominator  
  def set_nominator(self, nominator):
    self.nominator = nominator
  def set_denominator(self, denominator):
    self.denominator = denominator
  def __str__(self):
    return str(self.nominator/self.denominator)
  def __add__(self, other):
    return ((self.nominator * other.denominator) + (other.nominator * self.denominator)) / (self.denominator * other.denominator)
  def __sub__(self, other):
    return ((self.nominator * other.denominator) - (other.nominator * self.denominator)) / (self.denominator * other.denominator)

f1 = Fraction(3,2)
f2 = Fraction(2,3)
print(f1-f1)
```

```python
class Food(object):
  def __init__(self, n, v, w):
    self.name = n
    self.value = v
    self.calories = w
  def getValue(self):
    return self.value
  def getCost(self):
    return self.calories
  def density(self):
    return self.getValue()/self.getCost()
  def __str__(self):
    return self.name + ': <' + str(self.value)\
    + ', ' + str(self.calories) + '>'
```

```python
def buildMenu(names, values, calories):
  """names, values, calories lists of same length.
  name a list of strings
  values and calories lists of numbers
  returns list of Foods"""
  menu = []
  for i in range(len(values)):
    menu.append(Food(names[i], values[i], calories[i]))
  return menu
```

```python
def greedy(items, maxCost, keyFunction):
  """Assumes items a list, maxCost >= 0,
  keyFunction maps elements of items to numbers"""
  itemsCopy = sorted(items, key = keyFunction, reverse = True)
  result = []
  totalValue, totalCost = 0.0, 0.0
  for i in range(len(itemsCopy)):
    if (totalCost+itemsCopy[i].getCost()) <= maxCost:
      result.append(itemsCopy[i])
      totalCost += itemsCopy[i].getCost()
      totalValue += itemsCopy[i].getValue()
  return (result, totalValue)
```

```python
def testGreedy(items, constraint, keyFunction):
  taken, val = greedy(items, constraint, keyFunction)
  print('Total value of items taken =', val)
  for item in taken:
    print(' ', item) 
```

```python
def fib(n):
  if n == 0 or n == 1:
    return 1
  else:
    return fib(n - 1) + fib(n - 2)

fib(120)
```

```python
def fastFib(n, memo = {}):
  """Assumes n is an int >= 0, memo used only by
  recursive calls
  Returns Fibonacci of n"""
  if n == 0 or n == 1:
    return 1
  try:
    return memo[n]
  except KeyError:
    result = fastFib(n-1, memo) + fastFib(n-2, memo)
    memo[n] = result
    return result

fib(120)
```

```python
import random

def runSim(goal, numTrials, txt):
  total = 0
  for i in range(numTrials):
    result = ''
    for j in range(len(goal)):
      result += str(random.choice([1, 2, 3, 4, 5, 6]))
    if result == goal:
      total += 1
  print('Actual probability of', txt, '=', round(1 / (6 ** len(goal)), 8))
  estProbability = round(total / numTrials, 8)
  print('Estimated Probability of', txt, '=', round(estProbability, 8))
runSim('11111', 1000000, '11111')
```

```python
import math
def birthdayProp(numPeople):
  numerator = math.factorial(366)
  denom = (366**numPeople)*math.factorial(366-numPeople)
  return 1 - numerator/denom
```

```python
class Location(object):
  def __init__(self, x, y):
    """x and y are floats"""
    self.x = x
    self.y = y
  
  def move(self, deltaX, deltaY):
    """deltaX deltaY are floats"""
    return Location(self.x + deltaX, self.y + deltaY)
  
  def getX(self):
    return self.x
  
  def getY(self):
    return self.y

  def distFrom(self, other):
    xDist = self.x - other.getX()
    yDist = self.y - other.getY()
    return (xDist ** 2 + yDist ** 2) ** 0.5
    
  def __str__(self):
    return '<' + str(self.x) + ', ' \
    + str(self.y) + '>'
```

```python
class Walker(object):
  def __init__(self):
    """Assumes name is a str"""
    self.name = __name__
  
  def __str__(self):
    return self.name if self != None else 'Anonymous'
```

```python
import random


class UsualWalker(Walker):
  def takeStep(self):
    stepChoices = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    return random.choice(stepChoices)

class NorthernWalker(Walker):
  def takeStep(self):
    stepChoices = [(0.0, 1.1), (0.0, -0.9),
    (1.0, 0.0), (-1.0, 0.0)]
    return random.choice(stepChoices)
```

```python
class Field(object):
  def __init__(self):
    self.walkers = {}

  def addWalker(self, walker, loc):
    if walker in self.walkers:
      raise ValueError('Duplicate')
    else:
      self.walkers[walker] = loc
  
  def getLoc(self, walker):
    if walker not in self.walkers:
      raise ValueError('Not in the field')
    return self.walkers[walker]

  def moveWalker(self, walker):
    if walker not in self.walkers:
      raise ValueError('Not in the field')
    xDist, yDist = walker.takeStep()
    # use move method of Location
    # to get new location
    loc = self.walkers[walker]
    self.walkers[walker] = loc.move(xDist, yDist)
```
