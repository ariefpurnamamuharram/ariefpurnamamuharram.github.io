a. Write a Class (with the appropriate function(s)) to calculate a student’s final grade, based on the
average of their test scores. You may use any data structures to save the student’s grade, use the
most appropriate one.

```python
class NilaiAkhir(object):
  def __init__(self,L):         # L adalah list nilai dari siswa
    self.L = L
  def avg(self):
    sum = 0
    for i in self.L:
      sum = sum + i
    rata = sum/len(self.L)
    return(rata)
```

```python
# Example

Adriel = [80,81,82,83,84]
absen1 = NilaiAkhir(Adriel)

print(absen1.avg())
```

Explain the usage of these keywords and give an appropriate example for each keyword.<br>
a. __add__<br>
b. __eq__<br>
c. __str__<br>
d. self<br>
e. pass<br>

```python
# Keyword __add__() 
# ----
# Keyword __add__() is one of the special keywords in Python that defines two addition operator (+) between objects

class Tesla:
  def __init__(self, price):
    self.price = price
  def __add__(self, obj):
    return self.price + obj.price


car1 = Tesla(3500)
car2 = Tesla(3500)
total_price = car1 + car2

print(total_price)
```

```python
# Keyword __eq__() 
# ----
# Keyword __eq__() is one of the special keywords in Python that act as comparison check (==) between objects


class Car:
  def __init__(self, name):
    self.name = name
  def __eq__(self, obj):
    return self.name == obj.name

toyota = Car('Toyota')
tesla = Car('Tesla')

print(toyota == tesla)
```

```python
# Keyword __str__()
# ----
# Keyword __str__() is one of the special keywords in Python that return strings when the object is printed using print() function


class Car:
  def __init__(self, name):
    self.name = name
  def __str__(self):
    return "You have {car_name} in your garage!".format(car_name=self.name)

car = Car('Tesla')
print(car)
```

```python
# self
# ----
# Self is a parameter in Python that refer to the current instance of a class.
# Self parameter is used to access variables that defined in the class 


class Car:
  def __init__(self, name):
    self.name_by_self_parameter = name
  def get_name_by_self_parameter(self):
    return self.name_by_self_parameter
  def get_name_not_using_self_parameter(self):
    return 'Car'

c = Car('Tesla')
print(c.get_name_by_self_parameter())
print(c.get_name_not_using_self_parameter())
```
